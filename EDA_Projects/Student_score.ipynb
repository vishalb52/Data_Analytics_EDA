```python
student  = data.drop(['race/ethnicity', 'parental level of education'], axis=1)
```

```python
#relationship analysis
```

```python
corelation = student.corr()
```

```python
#Heatmap = A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. ... Correlation heatmaps are ideal for comparing the measurement for each pair of dimension values.
#xticks( ticks ) sets the x-axis tick values, which are the locations along the x-axis where the tick marks appear. ... Use this option if you want to retain the current tick values when resizing the axes or adding new data to the axes.Same for Ytickslabels
#annot – an array of same shape as data which is used to annotate the heatmap. cmap – a matplotlib colormap name or object. This maps the data values to the color space. fmt – string formatting code to use when adding annotations. linewidths – sets the width of the lines that will divide each cell
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True)
```
